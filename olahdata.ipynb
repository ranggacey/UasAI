```python
df = pd.read_csv('apk.csv')

total_penggunaan = df['Penggunaan'].sum()
df['Persentase'] = (df['Penggunaan'] / total_penggunaan) * 100


print(df[['Aplikasi', 'Persentase']])
```

```python


# Menyaring data hanya untuk aplikasi dengan penggunaan di atas 10 juta
high_usage_apps = df[df['Penggunaan'] > 10000000]
print(high_usage_apps)
```

```python

def kategori_pemakaian(penggunaan):
    if penggunaan > 50000000:
        return 'Tinggi'
    elif penggunaan > 20000000:
        return 'Sedang'
    else:
        return 'Rendah'

# Menambahkan kolom 'Kategori' berdasarkan jumlah penggunaan
df['Kategori'] = df['Penggunaan'].apply(kategori_pemakaian)

# Menampilkan lima baris pertama dari dataframe dengan kolom tambahan
print(df)
```

```python
import matplotlib.pyplot as plt

# Data maksimum (Shopee) dan minimum (Hijup)
maximum_usage = df[df['Aplikasi'] == 'Shopee']['Penggunaan'].values[0]
minimum_usage = df[df['Aplikasi'] == 'Hijup']['Penggunaan'].values[0]


apps = ['Shopee', 'Hijup']
usage = [maximum_usage, minimum_usage]

plt.figure(figsize=(8, 6))
plt.bar(apps, usage, color=['blue', 'green'])
plt.xlabel('Aplikasi')
plt.ylabel('Penggunaan')
plt.title('Perbandingan Penggunaan Maksimum (Shopee) dan Minimum (Hijup)')
plt.show()
```
